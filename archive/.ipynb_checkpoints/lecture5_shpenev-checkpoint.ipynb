```python
kard2 = kardashians.rename(
    columns = {
        'Kim Kardashian: (Worldwide)' : 'kim',
        'Khlo√© Kardashian: (Worldwide)' : 'khloe',
        'Kourtney Kardashian: (Worldwide)' : 'kourtney',
        'Kendall Jenner: (Worldwide)' : 'kendall',
        'Kylie Jenner: (Worldwide)' : 'kylie',
        }
)

kard2.columns = ["month", "kim", "khloe", "kourtney", "kendall", "kylie"]
display(kard2)


### R-Equivalent ###
# kard2 <- kardashians %>%
#   rename(month = Month,
#          kim = `Kim Kardashian: (Worldwide)`,
#          khloe = `Khlo√© Kardashian: (Worldwide)`,
#          kourtney = `Kourtney Kardashian: (Worldwide)`,
#          kendall = `Kendall Jenner: (Worldwide)`,
#          kylie = `Kylie Jenner: (Worldwide)`)
# colnames(kardashians) <- c("month", "kim", "khloe", "kourtney", "kendall", "kylie")
```

```python
kard3 = kard2.copy()

kard3['khloe'] = kard3['khloe'].apply(lambda x: x.replace('<1','0.5'))
kard3['kourtney'] = kard3['kourtney'].apply(lambda x: x.replace('<1','0.5'))
kard3['kendall'] = kard3['kendall'].apply(lambda x: x.replace('<1','0.5'))
kard3['kylie'] = kard3['kylie'].apply(lambda x: x.replace('<1','0.5'))

kard3['kim'] = pd.to_numeric(kard3['kim'])
kard3['khloe'] = pd.to_numeric(kard3['khloe'])
kard3['kourtney'] = pd.to_numeric(kard3['kourtney'])
kard3['kendall'] = pd.to_numeric(kard3['kendall'])
kard3['kylie'] = pd.to_numeric(kard3['kylie'])

# #OR, a shortcut:
kard3_short = kard2.copy().replace({'<1' : 0.5}, inplace=False)
kard3_short[['khloe','kourtney','kendall','kylie']] = kard3_short[['khloe','kourtney','kendall','kylie']].apply(pd.to_numeric)

### R-Equivalent ###
# kard3 <- kard2 %>%
#   mutate(khloe = str_replace(khloe, "<", ""),
#          kourtney = str_replace(kourtney, "<", ""),
#          kendall = str_replace(kendall, "<", ""),
#          kylie = str_replace(kylie, "<", "")) %>% 
#   mutate(khloe = as.numeric(khloe),
#          kourtney = as.numeric(kourtney),
#          kendall = as.numeric(kendall),
#          kylie = as.numeric(kylie))

# kard3 
```

```python
from datetime import datetime, timedelta

kard4 = kard3.copy()

kard4['dt_date'] = kard4['month'].apply(lambda x: datetime.strptime(x, '%Y-%m') + timedelta(days=14))
kard4['dt_year'] = kard4['dt_date'].apply(lambda x: x.year)
kard4['dt_month'] = kard4['dt_date'].apply(lambda x: x.month)
kard4

### R-equivalent ###
# kard4 <- kard3 %>% 
#   separate(month, into = c("year", "month"), sep = "-", convert = TRUE) %>%
#   mutate(day = 15, .after = month) %>%
#   mutate(date = ymd(paste(year, month, day, sep="-")))
```

```python
kard4[kard4.dt_year == 2020]

### R-equivalent ###
#   kard4 %>%
#     filter(year == 2020)


kard4[kard4.dt_year == 2020][['kim','khloe','kourtney','kendall','kylie']].mean()

### R-equivalent ###
#   kard4 %>%
#     filter(year == 2020) %>% 
#     summarize(kim = mean(kim),
#               khloe = mean(khloe),
#               kourtney = mean(kourtney),
#               kendall = mean(kendall),
#               kylie = mean(kylie))
```

```python
kard5 = kard4.copy().groupby('dt_year').mean().iloc[:,0:5]
kard5
```

```python
#Who was most/least popular?
poi = 'khloe'

#When was kim most popular?
display(kard5.sort_values(by=poi, ascending=False))
#OR
display(kard5.nlargest(1, columns=poi))


#When was kim least popular?
display(kard5.sort_values(by=poi, ascending=True))
#OR
display(kard5.nsmallest(1, columns=poi))


### R-Equivalent ###
# kard5 %>% arrange(kim)
# kard5 %>% arrange(-kim)
# kard5 %>% top_n(1, kim)
```

```python
# Who is more popular? The Kardashians or the Jenners?
kard5['kardashians'] = (kard5['kim'] + kard5['khloe'] + kard5['kourtney']) / 3
kard5['jenners'] = (kard5['kendall'] + kard5['kylie']) / 2
versus_df = kard5[['kardashians','jenners']].reset_index()
versus_df['dt_year'] = versus_df['dt_year'].astype('str')

plt.figure(figsize = (15, 5))
plt.plot('dt_year', 'kardashians', data=versus_df)
plt.plot('dt_year', 'jenners', data=versus_df, color='red')
### R-equivalent ###
# ggplot(kard5,
#        aes(x = year,
#            y = kim)) +
#   geom_line()

# plot(kard5$year, kard5$kim, type = "l")
```

```python
kard_long = pd.melt(
    kard5, 
    value_vars = ['kim','khloe','kourtney','kendall','kylie'], 
    var_name = 'name',
    value_name = 'popularity',
    ignore_index = False
).reset_index()

display(kard_long.head())

import seaborn as sns
sns.set_theme(style="darkgrid")

plt.figure(figsize = (15, 5))
graph = sns.lineplot(y = 'popularity', x = 'dt_year', hue = 'name', data = kard_long)
graph.xaxis.set_major_locator(MultipleLocator(3))
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Keeping up with the Kardashians')
plt.show()
```
