### 1. Loading data into Jupyter Notebook 

```python
#import all necessary libraries for data cleaning and visualization
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style(style="whitegrid")
```

```python
#convert the loaded data and save it as a panda's data frame
crx_data = pd.read_csv("crx.data.csv",
                       index_col=False,
                       names=["Male", "Age", "Debt","Married", "BankCustomer","EducationLevel","Ethnicity","YearsEmployed",
                              "PriorDefault","Employed","CreditScore","DriversLicense","Citizen","ZipCode","Income","Approved"])
crx_data.head()
```

Before continuing our analysis, it is a good practice to make a copy of your original data and work on the copied version. We do this to avoid any potential rewriting of the original data.

```python
#make a copy and take a look on each column data type
crx_data_clean = crx_data.copy()
crx_data_clean.info()
```

### 2. Data cleaning, manipulation, and visualization

From the result above, there are 15 columns, including one dependent variable called "Approved." However, some of the columns seem to have the wrong data type; for example, the variable age is considered an object or string instead of a float data type. Before we start changing all the columns' data types to the correct ones, we should first find out whether there are missing values in any of our columns.

```python
#check NA
print(crx_data_clean.isnull().sum())

#check value format for variable 'Age'
print(crx_data_clean.Age.value_counts())
```

Python does not identify any missing values (NaN) in any of the columns. However, if we take a look at the values of the 'Age' variable, we can see that there are 12 rows of data that contains '?' instead of NaN or age number. We will not be able to convert this variable data type into float right now since there is a '?' value considered a string. Before we convert the '?' into NaN, we should take a look at whether the other columns have the same problem or not. 

```python
#the number of columns that contain '?'
crx_data_clean[crx_data_clean == '?'].count()
```

It looks like our intuition is correct! Multiple columns contain '?' as one of the values. Now that we have identified all the '?' values, we can transform them into a NaN before filling them with other values to run our machine learning models. 

```python
#change all '?' into NaN
cols = ['Male','Age','Married','BankCustomer','EducationLevel','Ethnicity','ZipCode']
crx_data_clean[cols] = crx_data_clean[cols].replace({'?':np.nan})
print(crx_data_clean.isnull().sum())
```

Since we have transformed all '?' values into NaN, we can start filling these missing values using several methods. We can use the mode approach to fill the missing values for categorical variables, 'Male', 'Married', 'BankCustomer', 'EducationLevel', 'Ethnicity', and 'ZipCode'. Succinctly, this approach will fill the missing values with each column's mode value. 

```python
cols2 = ['Male','Married','BankCustomer','EducationLevel','Ethnicity','ZipCode']
crx_data_clean[cols2]=crx_data_clean[cols2].fillna(crx_data_clean.mode().iloc[0])
print(crx_data_clean.isnull().sum())
```

Next, we need to transform our 'Age' column data type into float. One other column that we need to convert is the 'Approved' column, our dependent variable. The 'Approved' column contains '+' and '-' as its value instead of 1 and 0. Therefore, we need to change '+' into 1s and '-' into 0s for the machine learning models to identify them. 

```python
#change the "Age" column from string to float
crx_data_clean['Age'] = crx_data_clean['Age'].astype(float)

#change the "Approved" column value, '+' into 1, '-' into 0, and "Male" column value to 1 & 0
crx_data_clean['Approved'] = crx_data_clean['Approved'].map({'+': 1, '-': 0})

#double check to see if we have succesfully change the data types
crx_data_clean.info()
```

Now we can fill in 'Age' missing values. There are a couple of approaches that we can use to do this, e.g., mean, median, or linear regression. Since we have a lot of interconnecting variables, I believe we can try to predict each customer's missing age by looking at the trend of the other customers. Therefore, I believe using linear regression to predict the missing age values is appropriate for this dataset. There is one caveat: we cannot fit categorical data in string format. We will convert the categorical values into numerical data in the next section, so for now, we can try to predict the missing age based on the other continuous variables. 

```python
#create a df that contains only the continuous variables
crx_data_continous = crx_data_clean.select_dtypes(exclude=['object']).copy()
crx_data_continous.head()
```

```python
print(crx_data_continous.isnull().sum())
```

```python
#fill missing values (age = 0) using linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

#create test and train df for age
test_age = crx_data_continous.loc[crx_data_continous['Age'].isnull()]
train_age = crx_data_continous.loc[crx_data_continous['Age'].notnull()]

#take the age column from the train_age df and assign it as the dependent variable (y)
y_age = train_age['Age']

#then drop the age column
train_age.drop('Age',axis=1,inplace=True)
#now you have dependent variable = y, and independent variables = the rest of the continous variables

#fit both independent and dependent variables to the lr model
lr.fit(train_age,y_age)

#drop age column from the test df
test_age.drop('Age',axis=1,inplace=True)

#predict test_age
pred = lr.predict(test_age)

#assign the predicted value to the test df
test_age['Age'] = pred

#check test df to see the predicted value
print(test_age)
```

```python
#assign the new predicted age values back to the original df
crx_data_clean.loc[test_age.index,'Age'] = test_age['Age']

#double check if NAs still exist
crx_data_clean.isnull().sum()
```

Next, let's work on our categorical variables a bit more. To fit our data into machine learning models, we need to convert all categorical values into numeric. Additionally, we need to drop the 'ZipCode' column as this variable does not have any predictive ability towards customers' financial ability and the likelihood of their application getting accepted. Moreover, this column has a lot of different values that will decrease the accuracy of our model if not removed. 

We will convert all of the categorical columns using a scikit module called **OneHotEncode**. This module will create a dummy value of 1 and 0 for each combination of columns and values. One downside to this approach is that it will create additional columns, as many as possible combinations.

```python
#drop zip code from the existing data frame
crx_data_clean.drop('ZipCode', axis=1, inplace=True)

#create a copy of our existing dataframe that will function as our reference table
categorical_reference = crx_data_clean.copy()
categorical_reference = categorical_reference[['Male','Married','BankCustomer','EducationLevel','Ethnicity','PriorDefault','Employed','DriversLicense','Citizen']]
```

```python
from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
#passing all categorical columns into the encoder
enc_df = pd.DataFrame(enc.fit_transform(categorical_reference).toarray())
#merge with the reference table on key values
categorical_reference = categorical_reference.join(enc_df)
categorical_reference.head()
```

```python
#join the encoded columns to the original table
crx_data_clean = crx_data_clean.join(enc_df)
crx_data_clean.head()
```

```python
#final check before moving on to fit the data on machine learning models
crx_data_clean.isnull().sum()
```

### 3. Data Split and Standardization

Since we are all done cleaning our data, we can move forward with fitting the data into machine learning models. However, since all of our data have different ranges, we need to standardize them to reduce errors and biases. Let's first split the data into our train and test data frame. Once that is done, we can then standardize our dataset.

```python
#remove categorical variables that contains string value
crx_data_final = crx_data_clean.select_dtypes(exclude=['object']).copy()

#split dataset into train and test, and set the random_state so it maintain the same result
from sklearn.model_selection import train_test_split

data_features = crx_data_final.loc[:, crx_data_final.columns!='Approved']
data_target = crx_data_final['Approved']

x_train, x_test, y_train, y_test = train_test_split(data_features,
                                                    data_target,
                                                    test_size = 0.2,
                                                    random_state = 808)
```

```python
#standardize the data
from sklearn.preprocessing import StandardScaler

data_features = StandardScaler().fit_transform(data_features)
print(data_features)
```

### 4. Comparing multiple machine learning models

```python
#Decision tree
from sklearn import tree

model_clf = tree.DecisionTreeClassifier(random_state=101)
fitted_model_clf = model_clf.fit(x_train,y_train)
test_predictions_clf = fitted_model_clf.predict(x_test)
accuracy_clf = fitted_model_clf.score(x_test,y_test)
print(accuracy_clf)
```

```python
#create an array of column names to be included in the tree plot
feature_columns = crx_data_final.loc[:, crx_data_final.columns!='Approved'].columns.tolist()
print(feature_columns)

#Plot tree from decision tree
plt.figure(figsize=(35,10))
tree.plot_tree(fitted_model_clf,
          filled=True,
          rounded=True,
          class_names=["Approved","Rejected"],
          feature_names=feature_columns,
          fontsize=8,
          max_depth=6)
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000)
fitted_model_lr = model_lr.fit(x_train,y_train)
test_predictions_lr = fitted_model_lr.predict(x_test)
accuracy_lr = fitted_model_lr.score(x_test,y_test)
print(accuracy_lr)
```

```python
#Bagging
from sklearn.ensemble import BaggingClassifier
model_bag = BaggingClassifier(
    base_estimator = tree.DecisionTreeClassifier(),
    n_estimators = 800,
    max_samples = 0.8,
    oob_score = True,
    random_state = 801
)
fitted_model_bag = model_bag.fit(x_train,y_train)
test_predictions_bag = fitted_model_bag.predict(x_test)
accuracy_bag = fitted_model_bag.score(x_test,y_test)
print(accuracy_bag)
print(model_bag.oob_score_)
```

```python
#Boosting
from xgboost import XGBClassifier

model_xgboost = XGBClassifier(
    n_estimators = 1000, 
    learning_rate = 0.05,
    eval_metric='logloss',
    use_label_encoder=False)
fitted_model_xgboost = model_xgboost.fit(x_train,y_train, early_stopping_rounds = 5, eval_set = [(x_test, y_test)],verbose = False)
test_predictions_xgboost = fitted_model_xgboost.predict(x_test)
accuracy_xgboost = fitted_model_xgboost.score(x_test,y_test)
print(accuracy_xgboost)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier(max_features='sqrt',
                                   random_state=201,
                                   oob_score=True)
fitted_model_rfc = model_rfc.fit(x_train,y_train)
test_predictions_rfc = fitted_model_rfc.predict(x_test)
accuracy_rfc = fitted_model_rfc.score(x_test,y_test)
print(accuracy_rfc)
```

```python
#SVM
from sklearn.svm import SVC

model_svm = SVC(kernel='linear')
fitted_model_svm = model_svm.fit(x_train,y_train)
test_predictions_svm = fitted_model_svm.predict(x_test)
accuracy_svm = fitted_model_svm.score(x_test,y_test)
print(accuracy_svm)
```

```python
#Passive Aggresive Classifier
from sklearn.linear_model import PassiveAggressiveClassifier

model_pac = PassiveAggressiveClassifier(C=0.5,random_state=301)
fitted_model_pac = model_pac.fit(x_train,y_train)
test_predictions_pac = fitted_model_pac.predict(x_test)
accuracy_pac = fitted_model_pac.score(x_test,y_test)
print(accuracy_pac)
```

```python
#Radius Neighbors Classifier
from sklearn.neighbors import RadiusNeighborsClassifier

model_rn = RadiusNeighborsClassifier(radius=4600)
fitted_model_rn = model_rn.fit(x_train,y_train)
test_predictions_rn = fitted_model_rn.predict(x_test)
accuracy_rn = fitted_model_rn.score(x_test,y_test)
print(accuracy_rn)
```

### 6. Brief overview on Passive Aggresive and Radius Neighbors Classifiers

In addition to the machine learning models we have studied in class, two other models are relatively new to us, Passive Aggressive and Radius Neighbors Classifiers. Below is the explanation of what these two models are:

1. **Passive Aggressive Classifiers**: This model is considered an online model, where it is most suitable to process a large stream of data such as Twitter data. This model works by taking a sample of data, analyzing it, adjusting the algorithm (if necessary), and throwing away said sample of data. The term *passive* means that when the model can perfectly predict all data, it will not change the algorithm. In contrast, the term *aggressive* means that when the model fails to predict the outcome variable just in the slightest, it will change the algorithm to compensate for the failed prediction for every set of a new sample of data. 
2. **Radius Neighbors Classifiers**: This model is similar to K-Nearest Neighbors (KNN) model in concept. It will plot the new data in the model, and it will be classified depending on the highest number of neighboring data categories. However, instead of using a number of neighbors or data closest to the new data, this model utilizes a radius from the new data to identify its classification.
